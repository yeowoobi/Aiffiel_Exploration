```python

# (1) 필요한 모듈 import
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report




# (2) 데이터 준비
digits = load_digits()
digits_data = digits.data
digits_label = digits.target
```

scikit=learn에서 제공하는 `train_test_split`는 주어진 데이터를 학습 데이터와 테스트 데이터로 나눌 수 있는 함수를 말한다. <br>여기서 `test_size = 0.2` 는 학습 데이터와 테스트 데이터의 비를 $8:2$ 로 하여 학습 및 테스트를 진행하겠다는 뜻이다.

머신러닝은 크게 **지도학습(Supervised Learning)**과, **비지도학습(Unsupervised Learning)**의 두가지로 구분이 된다.
<br><br>


**지도학습**은 정답이 있는, 즉 사용자가 분류하고자 하는 라벨링이 명확하게 있는 문제에 대한 학습을 가리킨다. <br>이번 프로젝트 역시 분류하고자 하는 숫자가 명확하게 존재하므로 지도학습이라고 할 수 있다. <br><br>
반면 **비지도학습**은 정답이 없는 문제를 학습하는 것으로, 인공지능이 데이터들 간의 공통점을 찾아 스스로 분류하는 학습을 의미한다.

### 지도학습의 종류
---
지도학습은 다시 분류(Classification)와 회귀(Regression)으로 나뉘는데, 분류는 입력받은 특정 카테고리 중 하나로 분류하는 문제를 의미하고 <br>
회귀는 입력받은 데이터에 따라 특정 필드의 수치를 맞히는 문제를 말한다.
<br>
이 중, 이번 프로젝트에서 사용되는 지도학습은 __분류__ 이다.

### 분류 알고리즘
#### 프로젝트에 사용되는 알고리즘을 알아보자
---
#### Decision Tree Classifier
    - Decision Tree, 의사결정나무를 이용한 방법으로, 데이터를 분석하여 이들 사이에 존재하는 패턴을 예측 가능한 규칙들의 조합으로 나타내는 방법.
    - 의사결정나무 자체는 분류와 회귀 모두 가능. 즉, 범주나 연속형 수치 모두 예측이 가능하다는 의미.
    - 결정 경계가 데이터에 수직이라 특정 데이터에만 잘 작동할 가능성이 존재.


#### Random Forest Classifier
    - Random Forest는 Decision Tree 모델 여러개를 합쳐 Decision Tree의 단점을 보완한 모델.
    - Decision Tree를 만드는 데 있어 쓰이는 요소를 무작위로 선정하여 여러개의 Decision Tree를 생성.
    

#### SVM Classifier
    - Support Vector와 Hyperplane을 이용하여 분류를 수행하게 되는 선형 분류 알고리즘.
    - 2차원 공간에서 2개의 클래스만 존재할 때, 이 둘을 구별할 수 있는 선을 그어 구별하는 방법.
    - 이를, 모든 클래스와 구별하게하여 일대다 방법으로 사용해서 분류할 수 있음.

#### SGD Classifier
    - 배치크기가 1인 경사하강법 알고리즘. 
    - 반복이 충분하면 효과는 있지만 노이즈가 심한 단점이 있음.

#### Logistic Regression
    - 가장 널리 알려진 선형 분류 알고리즘으로, 소프트맥스 함수를 사용하여 다중 클래스 분류를 함.
    - 이름은 Regression(회귀) 이지만 실제로 수행하는 알고리즘은 Classification(분류)

### Decision Tree Classifier

```python
# DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier

# (4) 모델 학습 및 예측
decision_tree = DecisionTreeClassifier(random_state=32)
decision_tree.fit(x_train, y_train)
y_pred = decision_tree.predict(x_test)

print(classification_report(y_test, y_pred))
```

### Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

random_forest = RandomForestClassifier(random_state=2)
random_forest.fit(x_train, y_train)
y_pred = random_forest.predict(x_test)

print(classification_report(y_test, y_pred))
```

### SVM Classifier

```python
from sklearn import svm
svm_model = svm.SVC()
svm_model.fit(x_train, y_train)
y_pred = svm_model.predict(x_test)

print(classification_report(y_test,y_pred))
```

### SGD Classifier

```python
from sklearn.linear_model import SGDClassifier
sgd_model = SGDClassifier()
sgd_model.fit(x_train, y_train)
y_pred = sgd_model.predict(x_test)

print(classification_report(y_test,y_pred))
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression()

logistic_model.fit(x_train, y_train)
y_pred = logistic_model.predict(x_test)
print(classification_report(y_test, y_pred))
```

|:---:|:---:|

## 2. 와인분류

### SVM Classifier

```python
from sklearn import svm
svm_model = svm.SVC()
svm_model.fit(x_train, y_train)
y_pred = svm_model.predict(x_test)

print(classification_report(y_test,y_pred))
```

### SGD Classifier

```python
from sklearn.linear_model import SGDClassifier

sgd_model = SGDClassifier()
sgd_model.fit(x_train, y_train)
y_pred = sgd_model.predict(x_test)

print(classification_report(y_test,y_pred))
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression()

logistic_model.fit(x_train, y_train)
y_pred = logistic_model.predict(x_test)
print(classification_report(y_test, y_pred))
```

위 표를 참고하면, 같은 모델이라도 당연히 데이터셋에 따라 정확도의 차이가 존재하며 Model들 간에도 데이터셋에따라 정확도의 차이가 클 때도 있고, 작을 때도 있는 것을 볼 수 있다. 또한 A라는 데이터셋 에서는 정확도가 더 높았던 모델이 B라는 데이터 셋 에서는 정확도가 다른 모델들에 비해 떨어질 수도 있다는 것을 볼 수 있다.

단, 모델을 평가하는데 있어 정확도로만 판단하기에는 문제가 있을 수 있다. 예를들어 데이터셋이 불균형할 때, 경우에 따라 정확도에 큰 영향을 미쳐 실제 모델의 성능과는 다른 결과가 나올 가능성도 존재한다. 여러가지 클래스 중에 특정 클래스에 대하여 다른 클래스에 비해 정답률이 높을 가능성이 있을 수 있다는 의미이다.

또한 때에 따라 모델이 얼마나 정답을 맞췄느냐 보다 __모델이 얼마나 오답을 내지 않느냐__가 더 중요할 수도 있다. 손글씨 분류와 와인분류의 경우 클래스를 구별하는 것의 오답률과 정답률의 중요도가 비슷할 수 있지만 유방암진단 프로젝트의 경우에 암에 걸린 사람을 걸렸다고 진단하는 것 보다, 암에 걸린 사람을 걸리지 않았다고 판단하는 것이 더 중요할 것이다. <br><br>
이에 대한 성능을 비교하기 위하여 우리는 Precision(정밀도), Recall Sensitivity(재현율), f1 Score 등을 지표로 사용한다. 이 지표들은 모두 오차행렬의 True Positive, True Negative, False Positive, False Negative 을 이용한 지표들이다. <br>

$ Precision = \frac{TP} {TP+FP} $ <br><br>
$ Recall = \frac{TP} {TP+FN} $ <br><br>
$ f1 Score = 2\times\frac{(Precision)(Recall)} {Precision+Recall} $ <br><br>
$ Accuracy = \frac{TP+TN} {TP+TN+FP+FN} $ <br><br>

즉, 손글씨 분류와 와인분류의 경우에는 Accuracy 가, 유방암 진단의 경우에는 Recall 이 더 중요하다고 할 수 있다.
